```r
# load libraries
library(tidyverse)
library(tidymodels)
library(tibble)
```

```r
# read data
players_data <- read_csv("https://drive.google.com/uc?export=download&id=1Mw9vW0hjTJwRWx0bDXiSpYsO3gKogaPz")
table(players_data['experience'])
```

```r
 players_data  <- players_data |>  mutate(category_numeric = case_when(
    experience == "Pro" ~ 1,
    experience == "Regular" ~ 2,
    experience == "Beginner" ~ 3,
     experience == "Amateur" ~ 4,
     experience == "Veteran" ~ 5,
  ))

players_data
```

```r
# select columns needed for analysis
players <- players_data |>
    select(played_hours, age)
head(players)
```

```r
set.seed(1)

# split data into training and testing
players_split <- initial_split(players, prop = 0.75, strata = played_hours)
players_train <- training(players_split)
players_test <- testing(players_split)
```

```r
set.seed(1)

# create recipe
players_recipe <- recipe(played_hours ~ age, data = players_train) |>
    step_scale(all_predictors()) |> # it is not necessary to standardize predictors as there is only one
    step_center(all_predictors())   # but we have included it for good practice

# create model specification
players_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
    set_engine("kknn") |>
    set_mode("regression")

# create 5-fold cross validation object
players_vfold <- vfold_cv(players_train, v = 5, strata = played_hours)

# create workflow
players_wkflw <- workflow() |>
    add_recipe(players_recipe) |>
    add_model(players_spec)

# create grid of possible k values
gridvals <- tibble(neighbors = seq(from = 1, to = 90, by = 2))

# tune the model and collect metrics
players_results <- players_wkflw |>
    tune_grid(resamples = players_vfold, grid = gridvals) |>
    collect_metrics() |>
    filter(.metric == "rmse")

# visualize metrics
plot_results <- players_results |>
    ggplot(aes(x = neighbors, y = mean)) +
        geom_point() +
        geom_line()
plot_results
```

```r
# find optimal k value
min_kval <- players_results |>
    filter(mean == min(mean))
min_kval
```

```r
set.seed(1)

kmin <- min_kval |> pull(neighbors)

# recreate model with k = 15
players_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = kmin) |>
    set_engine("kknn") |>
    set_mode("regression")

players_fit <- workflow() |>
    add_recipe(players_recipe) |>
    add_model(players_spec) |>
    fit(data = players_train)

# predict on the testing data
players_summary <- players_fit |>
    predict(players_test) |>
    bind_cols(players_test) |>
    metrics(truth = played_hours, estimate = .pred) |>
    filter(.metric == 'rmse')

players_summary
```

```r
age_prediction_grid <- tibble(
    age = seq(
        from = players |> select(age) |> min(),
        to = players |> select(age) |> max(),
        by = 10
    )
)

players_preds <- players_fit |>
  predict(age_prediction_grid) |>
  bind_cols(age_prediction_grid)

players_preds

plot_final <- ggplot(players, aes(x = age, y = played_hours)) +
  geom_point(alpha = 0.4) +
  geom_line(data = players_preds,
            mapping = aes(x = age, y = .pred),
            color = "steelblue",
            linewidth = 1) +
  xlab("Age in years") +
  ylab("Played hours") +
  ggtitle(paste0("K = ", kmin)) +
  theme(text = element_text(size = 12))

plot_final
```

The columns of `players.csv`  and `sessions.csv`, which contain information about players in PLAIcraft pertain to the following definitions:
- `experience` (character);  amount of experience a player has in Minecraft
- `subscribe` (logical); whether the player is subscribed to emails
- `hashedEmail` (character); players email in hashed (coded) form, unique identifier
- `played_hours` (double); number of hours played on the server
- `name` (character); player's pseudonym on PLAIcraft
- `gender` (character); player's gender
- `age` (double); player's age in years
- `individualId` (logical); not accessible: if individual has an id
- `organizationName` (logical); not accessible: if signup is from an organization
- `start_time` and `end_time` (character); date in dd/mm/yyyy format and time in 24-hour format when the player logged onto/off the server
- `original_start_time` and `original_end_time` (double); a time in UNIX timestamp format, logging the scheduled start/end time as number of seconds since January 1 1970 at UTC
